```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("1 - Sample - Superstore.csv", encoding='windows-1252')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] =  df['Order Date'].dt.year
```

```python
df.head(2)
```

## Histogram

```python
plt.figure(figsize=(6,5))
plt.hist(df['Sales'], bins=7)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.hist(df['Profit'], bins=7)
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

## Scatter Plot

relationship bw 2 variables

```python
plt.figure()
plt.scatter(df['Sales'], df['Profit'])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

## Pie Chart

```python
category_wise_orders = df.groupby('Category')['Order ID'].count()
category_wise_orders
```

```python
plt.figure()
plt.pie(category_wise_orders.values, labels=category_wise_orders.index)
plt.title("Category wise order count")
plt.show()
```

```python
category_wise_profit = df.groupby('Category')['Profit'].sum()
category_wise_profit
```

```python
plt.figure()
plt.pie(category_wise_profit.values, labels=category_wise_profit.index)
plt.title("Category wise profit")
plt.show()
```

```python
region_wise_sales = df.groupby('Region')['Sales'].sum()
region_wise_sales
```

```python
plt.figure()
plt.pie(region_wise_sales.values, labels=region_wise_sales.index)
plt.title("Region wise sales")
plt.show()
```

## Heat map

coorelation bw numerial values

```python
corr = df[['Sales', 'Profit', 'Discount', 'Quantity']].corr()
```

```python
corr
```

```python
plt.imshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.colorbar()
plt.show()
```
