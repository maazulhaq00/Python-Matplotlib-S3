```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
salesdata = {
    "Month": ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    "Sales": [123, 340, 100, 190, 300, 300, 400, 700, 300, 180, 230, 500]
}
```

```python
plt.plot(salesdata["Month"], salesdata["Sales"])
plt.title("Month wise sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
